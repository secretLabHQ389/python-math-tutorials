```python
import matplotlib.pyplot as plt

# import numpy as np
# y = m*np.array(x)+b

# y = mx + b used to draw a line

x = [-5,5] # considered a list type
m = -.3
b = .1


for xi in range(0,len(x)):
    y[xi] = m*x[xi] + b
    
print(x,y) # values of x1,y1 and x2,y2

plt.plot(x,y,label='y=%sx+%s' %(m,b)) #not points, x nd y coordinates
plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)
plt.legend()

axis = plt.gca()
plt.plot(axis.get_xlim(),[0,0],'k--')
plt.plot([0,0],axis.get_ylim(),'k--')

plt.show()
```

```python
import sympy.plotting.plot as symplot
from sympy.abc import x

y = x**2
p = symplot(y,show=False)
p.ylim = [-15,15]
p.xlim = [-15,15]
p[0].line_color = 'm'
p.title = 'This evil plot was prevented!'
p.show()
```

```python
x,a = sym.symbols('x a')

expr = a/x

n = 1
m = 2

p = symplot(expr.subs(a,n),(x,-5,5),show=False)
p[0].label = '$y=%s$' %sym.latex(expr.subs(a,n))

p2 = symplot(expr.subs(a,m),(x,-5,5),show=False)
# error below label: IndexError: list index out of range
# p[1].label = '$y=%s$' %sym.latex(expr.subs(a,n))
p.extend(p2)

p[0].line_color = 'r'
p.ylim = [-5,5]
p.legend = True
p.show()
```

```python
A = np.zeros((10,14))
print(np.shape(A))

for i in range(0,np.shape(A)[0]):
    for j in range(0,np.shape(A)[1]):
        A[i,j] = 3*i-4*j
        plt.text(j,i,int(A[i,j]),horizontalalignment='center',verticalalignment='center')
        
print(A)
plt.imshow(A)
plt.plot([0,3],[8,2],'r',linewidth=4)

plt.set_cmap('Purples')
plt.show()
```

```python
# "soft coding"
N = 10
C = np.zeros((N,N))

for i in range(0,N):
    for j in range(0,N):
        # i and j are displayed as subscripts in a Latex formula:
        C[i,j] = (-1)**(i+j) #PEMDAS
        
plt.imshow(C)
plt.set_cmap('gray')
plt.tick_params(labelleft=False,labelbottom=False)
plt.show()
print(C)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Polygon

y = np.array([ [2,2],[2.5,4],[3.5,1] ])
p = Polygon(y,facecolor='g',alpha=1)

y1 = np.array([ [1,1],[2,3],[3,1] ])
p1 = Polygon(y1,facecolor='r',alpha=.7)

fig,ax = plt.subplots()
ax.add_patch(p)
ax.add_patch(p1)
ax.set_ylim([0,4])
ax.set_xlim([0,4])
plt.show()
```
