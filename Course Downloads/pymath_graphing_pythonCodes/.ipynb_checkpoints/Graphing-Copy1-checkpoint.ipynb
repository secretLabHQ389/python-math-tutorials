```python
import sympy as sym

x = sym.symbols('x')
f = x**2-3*x

xran = range(-11,12)

#answer correct to this line:
for i in xran:
    # plt.plot(x[f(i)],y[f(i)],'o') didn't get instr. only y point is from formula
    plt.plot(i,f.subs(x,i),'o')
    
plt.xlabel('x')
plt.ylabel('f(x) = %s' %f)
plt.show()
```

```python
import matplotlib.pyplot as plt

# import numpy as np
# y = m*np.array(x)+b

# y = mx + b used to draw a line

x = [-5,5] # considered a list type
m = -.3
b = .1


for xi in range(0,len(x)):
    y[xi] = m*x[xi] + b
    
print(x,y) # values of x1,y1 and x2,y2

plt.plot(x,y,label='y=%sx+%s' %(m,b)) #not points, x nd y coordinates
plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)
plt.legend()

axis = plt.gca()
plt.plot(axis.get_xlim(),[0,0],'k--')
plt.plot([0,0],axis.get_ylim(),'k--')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

m = 0.7
b = -2
y = m*np.array(x)+b

a = -1.25
f = 0.75
c = a*np.array(x)+f

plt.plot(x,y,label='y=%sx+%s' %(m,b)) #not points, x nd y coordinates
plt.plot(x,c,label='y=%sx+%s' %(a,f)) #not points, x nd y coordinates
plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)
plt.legend()

axis = plt.gca()
plt.plot(axis.get_xlim(),[0,0],'k--')
plt.plot([0,0],axis.get_ylim(),'k--')

plt.show()
```

```python
x = [-5,5]
m = [.7,-1.25]
b = [-2,.75]

for i in range(0,len(m)):
    y = m[i]*np.array(x) + b[i]
    plt.plot(x,y,label='y = %sx+%s' %(m[i],b[i]))

plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)
plt.legend()
axis = plt.gca()
plt.plot(axis.get_xlim(),[0,0],'k--')
plt.plot([0,0],axis.get_ylim(),'k--') 

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = range(-3,4)
y = np.zeros(len(x))

print(y)

for i in range(0,len(x)):
    y[i] = 2 - x[i]**2
    
plt.plot(x,y,'s-')
plt.show()
```

```python
import sympy as sym
from sympy.abc import x

y = x**2
p = sym.plotting.plot(y,show=False)
p.ylim = [-15,15]
p.xlim = [-15,15]
p[0].line_color = 'm'
p.title = 'This evil plot was prevented!'
p.show()
```

```python
import sympy.plotting.plot as symplot
from sympy.abc import x

y = x**2
p = symplot(y,show=False)
p.ylim = [-15,15]
p.xlim = [-15,15]
p[0].line_color = 'm'
p.title = 'This evil plot was prevented!'
p.show()
```

```python
x,a = sym.symbols('x a')

expr = a/x

n = 1
m = 2

p = symplot(expr.subs(a,n),(x,-5,5),show=False)
p[0].label = '$y=%s$' %sym.latex(expr.subs(a,n))

p2 = symplot(expr.subs(a,m),(x,-5,5),show=False)
# error below label: IndexError: list index out of range
# p[1].label = '$y=%s$' %sym.latex(expr.subs(a,n))
p.extend(p2)

p[0].line_color = 'r'
p.ylim = [-5,5]
p.legend = True
p.show()
```

```python
#my answer was correct and needed a y axis limit to display legibly (w/out the label):
# import sympy.plotting.plot as symplot
# f,v,j, x = sym.symbols('f v j x')

# f = j/(x**2-j)

# for i in range(1,5):
#     symplot(f.subs(j,i),(x,-5,5),show=False)

# # pr.ylim = [-10,10]
# pr.show()

p = None
p = symplot(y.subs(a,1),(x,-5,5),show=False)
# 1 element with index of 0:
p[0].label = '$%s$' %sym.latex(y.subs(a,1))

for i in range(2,5):
    print(i)
    # 2nd element with index of 1:
    p.extend(symplot(y.subs(a,i),(x,-5,5),show=False))
    # another element with index of 1:
    # error- trying to access 3rd element of a 2 element obj (symplot, label)
    # causes element out of range error on p element (p is a list of a symplot and a label, a list of 2 objects)
    # solution is turn p[i] into p[i-1]
    p[i-1].label = '$%s$' %sym.latex(y.subs(a,i))
    
p.ylim = [-10,10]
p.legend = True
p.show()
```
