```python
#Mean and Standard Deviation
import numpy as np
# from numpy import random
import matplotlib.pyplot as plt

desired_mean = 15
desired_std = 4.3

nums = np.random.randn(1000)
nums = nums - np.mean(nums) #removes randomness in mean and will always be 15
nums = nums/np.std(nums) #removes randomness in standard deviation

nums = nums*desired_std+desired_mean

plt.subplot(121) #1 row, 2 columns, this goes in 1st column
plt.plot(nums, 's')

plt.subplot(122)
plt.hist(nums)
plt.show()

print('This distribution has a mean of %s and a standard deviation of %s' \
      %(np.mean(nums),np.std(nums)))
```

```python
# Lect 101 ans-

def angle_convert_plot():
    #get input from user
    angle = eval(input('What angle to convert?')) #eval runs strings as #s
    unit = input('Which unit did you input (radian or degree)?')
    
    if unit[0] == 'r' or unit[0]=='R':
        rad = angle
        deg = np.rad2deg(angle)
    elif unit[0].lower() == 'd':
        deg = angle
        rad = np.deg2rad(angle)
    else:
        raise ValueError('Unknown unit!')
    
    #convert to <360
    deg = deg%360
    rad = rad%np.pi
    
    # make the graph
    plt.plot([0,1],[0,0],'r',linewidth=2)
    plt.plot([0,np.cos(rad)],[0,np.sin(rad)],'r',linewidth=3)
    plt.axis('square')
    plt.axis([-1,1,-1,1])
    plt.grid()
    plt.title('Angle is %s$^o$ or %s radians.' %(deg,rad))
    plt.show()

angle_convert_plot()
```

```python
import numpy as np
import matplotlib.pyplot as plt

mvects = 200
ms = np.random.rand(mvects)
ks = np.random.rand(mvects)*2*np.pi

for i in range(0,mvects):
    
    r = np.random.rand()
    
    if r < .4:
        colorP = [1,.2,.7]
    elif r > .4 and r < .8:
        colorP = [.7,.2,1]
    else:
        colorP = [0,1,0]
    
    plt.polar([0,ks[i]],[0,ms[i]],color=colorP)
   
plt.axis('off')
plt.show()
```
