```python
import numpy as np
import matplotlib.pyplot as plt

a = 1 #1/4p
h = -1
k = -2
n = 100

x = np.linspace(h-5,h+5,n)
y = a*(x-h)**2 + k

plt.plot(x,y)
plt.grid()
plt.axis('square')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

a = 1 #1/4p
p = 1/(4*a)
h = -1
k = -2
n = 100

x = np.linspace(h-2,h+2,n)
y = a*(x-h)**2 + k

#parabola
plt.plot(x,y,label='parabola')
#vertix
plt.plot(h,k,'ro',label='vertix')
#focus
plt.plot(h,k+p,'go',label='focus')
#directrix
d = k-p
plt.plot(x[[0,-1]],[d,d],'o-')
plt.grid()
plt.legend()
plt.axis('square')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

h = 2
k = -3
r = 3

axlim = r + np.max((abs(h),abs(k)))
x = np.linspace(-axlim,axlim,100)
X,Y = np.meshgrid(x,x)

Fxy = (X-h)**2+(Y-k)**2 - r**2
                   
plt.contour(X,Y,Fxy,0)
plt.plot([-axlim,axlim],[0,0],'k--')
plt.plot([0,0],[-axlim,axlim],'k--')
plt.plot(h,k,'go')

plt.gca().set_aspect('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

h = 1
k = 2
a = 2
b = 3

axlim = np.max((a,b)) + np.max((abs(h),abs(k)))
x = np.linspace(-axlim,axlim,100)
X,Y = np.meshgrid(x,x)

#function-
ellipse = (X-h)**2/a**2 + (Y-k)**2/b**2 - 1

plt.gca().set_aspect('equal')
plt.contour(X,Y,ellipse,0)
plt.plot(h,k,'go')
```

```python
import numpy as np
import matplotlib.pyplot as plt

a = 1
b = .5
h = 1
k = 2

axlim = 2*np.max((a,b)) + np.max((abs(h),abs(k)))
x = np.linspace(-axlim,axlim,100)
X,Y = np.meshgrid(x,x)

hyperbola = (X-h)**2/a**2 - (Y-k)**2/b**2 - 1

plt.contour(X,Y,hyperbola,0)
plt.plot(h,k,'go')

plt.plot([-axlim,axlim],[0,0],'k--')
plt.plot([0,0],[-axlim,axlim],'k--')

plt.gca().set_aspect('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = 16
ab = np.linspace(1,5,n)

axlim = 8
x = np.linspace(-axlim,axlim,100)
X,Y = np.meshgrid(x,x)

for i in range(0,n):
    F1 = X**2/ab[i]**2 - Y**2/ab[i]**2 - 1
    plt.contour(X,Y,F1,0,colors=[i/n,0,i/n])
    
    F2 = -X**2/ab[i]**2 + Y**2/ab[i]**2 - 1
    plt.contour(X,Y,F2,0)
    
plt.axis('off')
plt.gca().set_aspect('equal')
plt.show()

```
