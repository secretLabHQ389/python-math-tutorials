```python
import numpy as np
from IPython.display import display,Math

a = 3
b = 4

print('Multiplication of logs:')
res1 = np.log(a*b)
res2 = np.log(a)* np.log(b)
res3 = np.log(a) + np.log(b)

display(Math('\\log(%g\\times%g) = %g' %(a,b,res1)))
display(Math('\\log(%g)\\times\\log(%g) = %g' %(a,b,res2)))
display(Math('\\log(%g) + log(%g) = %g' %(a,b,res3)))

display(Math('\\log(%g\\times%g) = \\log(%g) + log(%g)' %(a,b,a,b)))
display(Math('\\log(%g\\times%g) \\neq \\log(%g)\\times\\log(%g)' %(a,b,a,b)))

print('Division of logs:')
res1 = np.log(a/b)
res2 = np.log(a)/ np.log(b)
res3 = np.log(a) - np.log(b)

display(Math('\\log\\left(\\frac{%g}{%g}\\right) = %g' %(a,b,res1)))
display(Math('\\frac{\\log(%g)}{\\log(%g)} = %g' %(a,b,res2)))
display(Math('\\log(%g) - log(%g) = %g' %(a,b,res3)))

display(Math('\\log\\left(\\frac{%g}{%g}\\right) = \\log(%g) - log(%g)' %(a,b,a,b)))
display(Math('\\log\\left(\\frac{%g}{%g}\\right) \\neq \\frac{\\log(%g)}{\\log(%g)}' %(a,b,a,b)))

print('Exponents of logs:')
res11 = np.log(a**b)
res21 = b*np.log(a)
print(res11 - res21)
print(res11,res21)

display(Math('\\log(%g^{%g}) = %g' %(a,b,res11)))
display(Math('%g*(\\log(%g)) = %g' %(b,a,res21)))
```

```python
import numpy as np
import matplotlib.pyplot as plt

lst = [-4,5.5,1,2]
# lst[maxval] list indices must be integers or slices, not numpy.float64
maxvalidx = np.argmin(lst)
print(lst[maxvalidx])

x = np.linspace(0,2*np.pi,20)
fx = -(np.cos(x) + np.sqrt(x))
fmax = np.max(fx)
fmaxidx = np.argmax(fx)
print(fmax,fmaxidx)
plt.plot(x,fx,'bo-')
plt.plot(x[fmaxidx],fmax,'rs')
plt.show()
```

```python
x = np.linspace(-2,2,51)
fx = x**3 + x**4
fminidx = np.argmin(fx)
plt.plot(x,fx,'bs-')
plt.plot(x[fminidx],fx[fminidx],'rs')
plt.xlim([-1.2,.5])
plt.ylim([-.2,.2])
plt.show()
```

```python
#Even function
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5,20)

fEven = x**2
fEvenNeg = (-x)**2
negExpon = -x**2

plt.plot(x,fEven)
plt.plot(x,fEvenNeg, 'ro')
plt.plot(x,negExpon, '-')

plt.show()
```

```python
#Odd function
fOdd = x**3
fOddNeg = (-x)**3
fNegOdd = -fOdd

plt.plot(x,fOdd)
plt.plot(x,fOddNeg,'ro')
plt.plot(x,fNegOdd,'g-')
plt.show()
```
