```python
# Rose curves-
import numpy as np
import matplotlib.pyplot as plt

fig,ax = plt.subplots(3,3,figsize=(10,6))
row, col = np.indices((3,3))

for i in range(0,9):
    k = i/2
    print(k)
    t = np.linspace(-6*np.pi,6*np.pi,250)
    x = np.cos(k*t)*np.cos(t)
    y = np.cos(k*t)*np.sin(t)
    
    r = row.ravel()[i]
    c = col.ravel()[i]

    ax[r,c].plot(x,y,'k')
    ax[r,c].axis('square')
    ax[r,c].axis('off')
    ax[r,c].set_title('k=%s' %k)


# pl = list([0,.5,1,1.5,2,2.5,3,3.5,4])

# for i in pl:
#     t = np.linspace(0,4*np.pi,100)
#     x = np.cos(i*t)*np.cos(t)
#     y = np.cos(i*t)*np.sin(t)
#     pr = range(0,9)
#     for i in pr:
#         r = row.ravel()[i]
#         c = col.ravel()[i]
    
#     ax[r,c].plot(x,y)
#     ax[r,c].axis('off')
#     ax[r,c].axis('square')

# plt.plot(x,y,'m',linewidth=3)
# plt.axis('square')
# plt.axis('off')
# plt.show()
```

```python
# http://mathworld.wolfram.com/ Find a squircle-
import numpy as np
import matplotlib.pyplot as plt

# x**4 + y**4 = 1
x = np.linspace(-1,1,2001)
y = (1 - x**4)**(1/4)

plt.plot(x,y,'k',linewidth=2)
plt.plot(x,-y,'k',linewidth=2)
plt.axis('square')
plt.axis('off')
plt.show()
```

```python
aa = range(10,200)
for i in aa:
    t = np.linspace(6,i*np.pi,250)
    k = -3

    x = np.cos(t)*np.exp(t/k)
    y = np.sin(t)*np.exp(t/k)

    plt.plot(x,y,'r')
plt.axis('square')
plt.axis('off')

plt.show()
```

```python
N = 150
ks = np.linspace(-6,-2,N)

# t = np.linspace(3,i*np.pi,100)
t = np.linspace(6,10*np.pi,1000)

for i in range(N):
    x = np.cos(t)*np.exp(t/ks[i])
    y = np.sin(t)*np.exp(t/ks[i])
    plt.plot(x,y,color=[i/N,0,i/N])
    
plt.axis('square')
plt.axis('off')
plt.show()
```

```python
# The logistic map to chart chaos-
x = .5
n = range(1,501)
# r = [1,4] in 1000 steps
d = np.linspace(1,4,1000)
de = int((3/1000)*1000)
rr = range(0,de)
# x**(n+1)=r*x**n(1-x**n)

for i in n:
    for p in rr:
        h = (p*x**i)*(1-x**i)-x**(i+1)
        plt.plot(h,'bo')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# %matplotlib notebook

# r = 3
rs = np.linspace(1,4,1000)

N = 501
x = .5+np.zeros(N)
endcap = np.arange(round(N*.9),N)
# x[0] = .5

for ri in range(len(rs)):
    for n in range(N-1):
        x[n+1] = rs[ri]*x[n]*(1-x[n]) + .2*np.sin(2*np.pi*x[n])**2
        
    u = np.unique(x[endcap])
    r = rs[ri]*np.ones(len(u))  
    plt.plot(r,u,'k.',markersize=1)

plt.show()
```
