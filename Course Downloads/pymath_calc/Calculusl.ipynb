```python
# Tangent lines-

x = sym.symbols('x')

f = x**2
df = sym.diff(f)

xa = 1

fa = f.subs(x,xa)
dfa = df.subs(x,xa)

xx = np.linspace(-2,2,200)
f_fun = sym.lambdify(x,f)(xx)
#f_fun(xx)
f_fun
df_fun = sym.lambdify(x,df)(xx)
tanline = dfa * (xx-xa) + fa
plt.plot(xx,tanline,label='tangent')

    
plt.plot(xx,f_fun,label='f(x)')
plt.plot(xa,fa,'ro')

plt.axis('square')
plt.axis([-2,2,-.5,2])

ax = plt.gca()
plt.plot(ax.get_xlim(),[0,0],'k--') #Gaussian curve has tangent on y = 0
plt.plot([0,0],ax.get_ylim(),'k--')
plt.legend()
plt.show()
```

```python
def computetangent(f,fa,bnds):
    # defining the derivative and values at point xa
    df = sym.diff(f)
    fa = f.subs(x,xa)
    dfa = df.subs(x,xa)
    
    # evaluate tangent line
    xx = np.linspace(bnds[0],bnds[1],200)
    return dfa * (xx-xa) + fa
```

```python

x = sym.symbols('x')
f = x**2
xx = np.linspace(-2,2,200)
ffun = sym.lambdify(x,f)(xx)

# tanline = computetangent(f,1,xx[[0,-1]])

for a in xx:
    y = computetangent(f,a,xx[[0,-1]])
    plt.plot(xx,y,color=[abs(a)/3,abs(a)/4,abs(a)/2])

plt.plot(xx,ffun)
# plt.plot(xx,tanline)
plt.axis('square')
plt.axis([-2,2,-1,3])
plt.axis('off')
plt.show()
```

```python
# Lect 127 Finding critical points-

# empirical method
x = np.linspace(-5,5,1001) # keep resolution odd for critical points
# ^ called 'districtizing'
fx = x**2 * np.exp(-x**2)
dfx = np.diff(fx)/(x[1]-x[0])

localmax = find_peaks(fx)[0]
localmin = find_peaks(-fx)[0]

print('The critical points are ' + str(x[localmax]) + ' ' + str(x[localmin]))

plt.plot(x,fx)
plt.plot(x[0:-1],dfx)
plt.plot(x[localmax],fx[localmax],'ro')
plt.plot(x[localmin],fx[localmin],'bs')
plt.show()
```

```python
# Lect 128-
# a = range(range(0,2)/8)
# fx = x**2*sym.exp**(-a*x**2)

a,x = sym.symbols('a,x')

baseexpr = x**2 * sym.exp(-a*x**2)

arange = np.arange(0,2.25,.25)
print(arange)
xrange = np.linspace(-3,3,100)

# setup the figure
fig,ax = plt.subplots(1,2)

for ai in arange:
    fx = baseexpr.subs(a,ai)
    dfx = sym.diff(fx)
    critpnts = sym.solve(dfx)
    # print(critpnts)
    
    # plotting
    ax[0].plot(xrange,sym.lambdify(x,fx)(xrange))
    ax[1].plot(xrange,sym.lambdify(x,dfx)(xrange))
    
    # test for the critical points
    if 1 in critpnts:
        display(Math('\\Rightarrow %s \\text{ has a critical point at x=}1!!\\text{ Whooho!}' \
                    %sym.latex(fx)))
    elif 2 in critpnts:
        display(Math('\\Rightarrow %s \\text{ has a critical point at x=}2!!\\text{ Whooho!}' \
                    %sym.latex(fx)))
    else:
        display(Math('\\quad %s \\text{ has NO critical points at x=1 or x=2.}' \
                    %sym.latex(fx)))
    
ax[0].set_ylim([0,2])
ax[0].set_title('Function')
ax[0].plot([1,1],[0,2],'--',color='gray')
ax[0].plot([2,2],[0,2],'--',color='gray')

ax[1].set_ylim([-2,2])
ax[1].set_title('It\'s Derivative')
ax[1].plot([1,1],[-2,2],'--',color='gray')
ax[1].plot([2,2],[-2,2],'--',color='gray')

plt.show()
```

```python
# Area between two curves
from matplotlib.patches import Polygon

x = sym.symbols('x')

symf = x**2
symg = x

# convert those from sym.symbols into functions that can be called:
f = sym.lambdify(x,symf)
g = sym.lambdify(x,symg)

xx = np.linspace(-2,2,55)

# add patch
xpatch = np.linspace(0,1,100)
ypatch = np.vstack((g(xpatch),f(xpatch))).T
# print(ypatch) - points along line for patch

fig,ax = plt.subplots()
ax.add_patch( Polygon(ypatch,facecolor='k',alpha=.3))

plt.plot(xx,f(xx))
plt.plot(xx,g(xx),'r')

plt.legend(['$f(x)=%s$' %(sym.latex(symf)), '$g(x)=%s$' %(sym.latex(symg))])
#([-x,x,-y,y]):
plt.axis([-.25,1.25,-.5,1.5])
plt.show()
```
