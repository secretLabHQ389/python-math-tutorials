```python
G = [3,3]

#ravel method

pl = range(1,9)
for i in pl:
    t = np.linspace(i*np.pi,i*np.pi,250)
    a = t
    x = a*np.cos(t)**3
    y = a*np.sin(t)**3
    plt.plot(x,y,'m',linewidth=2)
    

plt.axis('off')
```

```python
fig,ax = plt.subplots(3,3)

for i in range(9):
    a = t**i
    t = np.linspace(a*np.pi,a*np.pi,250)
    x = a*np.cos(t)**3
    y = a*np.sin(t)**3
    
    ax[0,0].plot(x,y)
    ax[0,0].axis('off')
    ax[0,0].axis('square')
    
plt.show()
```
