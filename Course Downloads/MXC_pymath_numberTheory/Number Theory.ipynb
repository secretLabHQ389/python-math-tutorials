```python
def test4perfect(number):
    number = 6
    divisors = []
    for d in range(1,number):
        if number%d == 0:
            divisors.append(d)

    return sum(divisors)==number
```

```python
nums2sqrt = np.linspace(2,101,50)
niterations = np.arange(3,9)

err = np.zeros((len(niterations),len(nums2sqrt)))

# Loop over numbers to square root
# [ni,num] means index,i number in range
for ni,num in enumerate(nums2sqrt):
    # set the number to compute sqrt of
    Y = num
    # Loop over the number of iterations
    for ii,iters in enumerate(niterations):
        # Heron's algorithm:
        x = Y/3 # initial guess
        for n in range(iters): 
            x = (x + Y/x) / 2
        # get the error
        err[ii,ni] = abs(x-np.sqrt(Y)) # if it's not equal, that's the err
```

```python
plt.imshow(-np.log(err),aspect=10,extent=[nums2sqrt[0],nums2sqrt[-1],niterations[-1],niterations[0]])
plt.show()
```

```python
num2sqr = 13
startEst = np.linspace(-num2sqr,num2sqr,40)
niters = 5

# initialize a results matrix
sqrtAlgResults = np.zeros((len(startEst),niters+1))

# Loop over starting estimates
for idx,startnum in enumerate(startEst):
    
    # initialize start value
    x = [startnum]

    # Heron's algorithm:
    for n in range(niters): 
        newx = (x[n] + num2sqr/x[n]) / 2
        x.append( newx )
        
    sqrtAlgResults[idx,:] = x
    
```

```python
maxN = 10000

largestPrimeFact = np.zeros(maxN+1,dtype=int)

for i in range(2,maxN+1):
    largestPrimeFact[i] = np.max(sym.primefactors(i))

plt.plot(largestPrimeFact, 'k.', markersize=.5)
plt.xlabel('Number')
plt.ylabel('Largest prime factor')
plt.show()

largestPrimeFact
```

```python
# count number of occurrences of max prime factors

uniqueMaxPrimes = np.unique(largestPrimeFact)

counts = np.zeros(len(uniqueMaxPrimes),dtype=int)

for i,u in enumerate(uniqueMaxPrimes):
    counts[i] = np.sum(largestPrimeFact == u) #summing boolean 1 values
    
counts

plt.plot(uniqueMaxPrimes,counts,'k.')
plt.xlim([-2,500])
plt.title('\n\n\n\n\n%s is the modal max prime factor' %uniqueMaxPrimes[np.argmax(counts)])
plt.show()
print('\n\n\n\n\n\n\n\n\n\n\n\n\n\n')
```

```python
numberlist = np.arange(maxN+1)
len(numberlist[largestPrimeFact<=5])
```

```python
smoothCount = np.zeros(len(uniqueMaxPrimes))

for i,u in enumerate(uniqueMaxPrimes):
    smoothCount[i] = len(numberlist[largestPrimeFact<=u])
```

```python
plt.plot(uniqueMaxPrimes,smoothCount,'k.')
plt.xlim([-2,500])
plt.title('K-smooth number frequencies')
plt.xlabel('Number of numbers')
plt.ylabel('Smooth number')
plt.show()
```
